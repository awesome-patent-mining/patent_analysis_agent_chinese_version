```python
query = f"""
SELECT 
    `applicant`,
    GROUP_CONCAT(
        CONCAT(`country`, '(', `patent_count_by_country`, ')') 
        SEPARATOR ', '
    ) AS `office_and_count`,
    CAST(SUM(`patent_count_by_country`) AS UNSIGNED) AS `total_patent_count`
FROM (
    SELECT 
        `current_assignee` AS `applicant`,
        `patent_office` AS `country`,
        CAST(COUNT(*) AS UNSIGNED) AS `patent_count_by_country`
    FROM 
        `patent_info`
    WHERE 
        `current_assignee` IS NOT NULL
    GROUP BY 
        `applicant`, `country`
) AS subquery
GROUP BY 
    `applicant`
ORDER BY 
    `total_patent_count` DESC;
"""
```

```python
# 获取申请人及其专利数量统计
query = f"""
SELECT 
    `申请人`,
    GROUP_CONCAT(
        CONCAT(`国家`, '(', `分国专利数`, ')') 
        SEPARATOR ', '
    ) AS `受理局及数量`,
    CAST(SUM(`分国专利数`) AS UNSIGNED) AS `总专利数`
FROM (
    SELECT 
        `[标]当前申请(专利权)人` AS `申请人`,
        `受理局` AS `国家`,
        CAST(COUNT(*) AS UNSIGNED) AS `分国专利数`
    FROM 
        `patent_info`
    WHERE 
        `[标]当前申请(专利权)人` IS NOT NULL
    GROUP BY 
        `申请人`, `国家`
) AS subquery
GROUP BY 
    `申请人`
ORDER BY 
    `总专利数` DESC;
"""
```

```python
import sys
import os
# 获取当前工作目录的父目录
parent_dir = os.path.dirname(os.getcwd())
# 将父目录添加到Python路径
sys.path.append(parent_dir)
from research_agent.core.storage import PatentDatabase
import pandas as pd

# 创建数据库连接
db = PatentDatabase()
db.connect()

try:
    # 获取申请人及其专利数量统计
    query = f"""

        SELECT 
            `[标]当前申请(专利权)人` AS `申请人`,
            `受理局` AS `国家`,
            CAST(COUNT(*) AS UNSIGNED) AS `分国专利数`
        FROM 
            `patent_info`
        WHERE 
            `[标]当前申请(专利权)人` IS NOT NULL
        GROUP BY 
            `申请人`, `国家`
    """
    
    # 执行查询
    db.cursor.execute(query)
    
    # 获取列名
    columns = [desc[0] for desc in db.cursor.description]
    
    # 获取结果
    results = db.cursor.fetchall()
    
    # 打印查询信息
    print(f"执行的查询: {query}")
    print(f"查询结果行数: {len(results)}")
    print("\n查询结果:")
    
    # 打印列名
    print("\t".join(columns))
    print("-" * 80)
    
    # 打印数据
    for row in results:
        print("\t".join(str(value) for value in row))
    
    # 转换为DataFrame并显示
    df = pd.DataFrame(results, columns=columns)
    print("\nDataFrame格式:")
    print(df)
    
    # 保存到CSV文件
    df.to_csv('patent_analysis_results.csv', index=False, encoding='utf-8')
    print("\n结果已保存到 patent_analysis_results.csv")
    
finally:
    db.disconnect()
```
